```python
import polars as pl

df = pl.read_csv("train.csv")

print(df) #выводим общую информацию о тренировочном датасете

print(df["Pclass"].value_counts()) #Выводим информацию о пассажирах разных классов

print(df.group_by("Sex").agg((pl.col("Survived")).sum())) #Выводим колличество выживших мужчин и женщин

print(df.filter(df["Age"] > 44)) #Выводим информацию о пассажирах с возрастом больше 44 лет
```

```python
import pandas as pd
import numpy as np
import bottleneck as bn


df = pd.read_csv('train.csv') #Считаю данные из датасета из файла train.csv с помощью pandas

bn.nanmedian(df['Age']) #Считаю средний возраст с помощью bottleneck
```

```python
bn.nanstd(df['Age']) #Считаю стандартное отклонение с помощью bottleneck
```

```python
df['Fare_new'] = np.where(df['Pclass'] < 3, df['Fare'] * 1.3, df['Fare'] * 1.1) #Провожу манипуляции над столбцом Fare и записываю в новый столбец
df.to_csv('train_modified.csv', index=False)

print(df.info()) #Вывожу информацию
```

**Оптимизация типов pandas**

```python
df_h = pd.read_csv("Housing.csv") #Считываю датасет из файла Housing.csv с помощью pandas
df_h.info() #Вывожу инофрмацию

df_h.head(20) #Вывож первые 20 значений
```

```python
df_h.describe() #Вывожу описание
```

выводы:


1.   Типы данных были изменены на более оптимальные, так как int64 может содержать слишком большие числа, которых нет в нашем датасете. Наш датасет подходит под тип данных int8.
2.   Параметр цены был достаточно высок, поэтому его пришлось оставить в int32.
3.   Параметр area решил уложить в тип int16.
4.   Также были введены категории для object, так как в датасете отсутствует большое количество уникальных значений.




```python
df_h['price']=df_h['price'].astype(np.int32)
df_h['area']=df_h['area'].astype(np.int16)
df_h['bedrooms']=df_h['bedrooms'].astype(np.int8)
df_h['bathrooms']=df_h['bathrooms'].astype(np.int8)
df_h['stories']=df_h['stories'].astype(np.int8)
df_h['parking']=df_h['parking'].astype(np.int8)
df_h['mainroad']=df_h['mainroad'].astype("category")
df_h['guestroom']=df_h['guestroom'].astype("category")
df_h['basement']=df_h['basement'].astype("category")
df_h['hotwaterheating']=df_h['hotwaterheating'].astype("category")
df_h['airconditioning']=df_h['airconditioning'].astype("category")
df_h['prefarea']=df_h['prefarea'].astype("category")
df_h['furnishingstatus']=df_h['furnishingstatus'].astype("category")
df_h.info()
```

Изначально датасет занимал 55,5 КБ, после оптимизации он стал занимать 10.0 КБ
